```python
import torch
import os
import pathlib
from super_gradients.training import models
from super_gradients.common.object_names import Models
```

```python
# Uploading YOLO-NAS Pose
# using large model NAS Pose with COCO weights
model = models.get("yolo_nas_pose_l", pretrained_weights="coco_pose")
# If downloading model from existing weights:
# model = models.get("yolo_nas_pose_l", num_classes = 17, checkpoint_path="/content/yolo_nas_pose_l_coco_pose.pth")
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)

confidence = 0.3
```

```python
# Inference video
video_path = "/content/multipose.mp4"

output_video_path = pathlib.Path(video_path).stem + "-detections" + pathlib.Path(video_path).suffix
model.predict(video_path, conf=confidence, fuse_model = False).save(output_video_path)
```
